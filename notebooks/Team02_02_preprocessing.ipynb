## Simple Cleaning

- Duplicate Rows: No duplicate rows were found in the dataset, so none were removed.
- Data Leakage: No columns were identified as data leakage. All features represent user behavior during the session and would be known at the time of prediction.
- Impossible Values: No impossible or invalid values (such as negative durations or unrealistic entries) were found in the dataset.
- Target Encoding: The target variable `Revenue` was converted from boolean (True/False) to binary (1/0) for modeling purposes.

Overall, no rows or columns were removed during this step, and the dataset remains clean and consistent.

```python
X = df.drop('Revenue', axis=1)
y = df['Revenue']
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X, y,
    test_size=0.2,
    random_state=4950,
    stratify=y
)
```

## Train-Test Split

- The dataset was split into 80% training and 20% testing sets using a random state of 4950.
- Stratified sampling was used to preserve the class distribution of the target variable (`Revenue`).

### Dataset Sizes
- Training set: 9,864 rows
- Test set: 2,466 rows

### Class Distribution
The class ratios in both the training and test sets are very similar, confirming that stratification worked correctly. This ensures that both sets are representative of the original dataset, which is especially important given the class imbalance.
